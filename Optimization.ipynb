```python
import numpy as np
import matplotlib.pyplot as plt


# Define the function f(x)
def f(x):
    return abs(-(x**2) + 4)


# Generate x values from -3 to 3
x = np.linspace(
    -3, 3, 400
)  # You can adjust the number of points (400 in this case) for smoother or coarser plotting

# Calculate corresponding y values
y = f(x)

# Create the plot
plt.plot(x, y, label="f(x) = | -x^2 + 4 |")

# Add labels and title
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x) = | -x^2 + 4 |")

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
