Definition of Historic Data

```python
''' Input of Historic Passenger Number Data
Observed Aggregated Demand for System XX'''

Paxnum = []
Data = {1982: 8367699, 1983: 8661582, 1984: 9102355, 1985: 9546141, 1986: 9670366, 1987: 10534015, 1988: 11305215, 1989: 12231671, 1990: 12769757,
          1991: 12225542, 1992: 13119273, 1993: 13574085, 1994: 14573334, 1995: 15395406, 1996: 16276699, 1997: 18318360, 1998: 19326589, 1999: 20925667, 2000: 22675366,
          2001: 21012871, 2002: 17948058, 2003: 17024937, 2004: 17252906, 2005: 17884652, 2006: 19237216, 2007: 20739113, 2008: 22099233, 2009: 21926872, 2010: 22878251,
          2011: 24337954, 2012: 24802400, 2013: 24865138, 2014: 25477622, 2015: 26281228, 2016: 27666428, 2017: 29396094, 2018: 31113488, 2019: 31507692, 2020: 8341047,
          2021: 10234428, 2022: 22561132}
#print(Data[1986])
```

Extracton of Historic Data

```python
for number in Data.values():
    Paxnum.append(number)

#print(Paxnum)
```

Extraction of Annual Change Data

```python
Annual_change = []
for i in range(1, len(Paxnum)):
    Difference = Paxnum[i] - Paxnum[i-1]
    Annual_change.append(Difference)
Annual_change = np.array(Annual_change)
#print(Annual_change)
```

Extraction of Annual Change in %

```python
Annual_change_per100 = []
for i in range(1, len(Paxnum)):
    Percent = ((Paxnum[i] / Paxnum[i-1])-1)*100
    Annual_change_per100.append(Percent)
Annual_change_per100 = np.array(Annual_change_per100)
#print(Annual_change_per100)
```

## Plots
Specific Plots for Sample Data

```python
x = np.arange(1982, 2023)
y = np.array(Paxnum)
plt.plot(x, Paxnum)


plt.plot(x, y, label='Line Plot', color='blue', marker='o')
plt.grid(True)
plt.xlabel("Year")
plt.ylabel("Passenger Numbers")
plt.title("Demand for Zurich Airport")
plt.legend()  # Display legend
plt.figure()


x2 = x[1:]
plt.plot(x2, Annual_change, label='Line Plot', color='blue', marker='o')
plt.grid(True)
plt.xlabel("Year")
plt.ylabel("Annual Change of Passenger Numbers")
plt.title("Demand for Zurich Airport")
plt.legend()  # Display legend
plt.figure()

x2 = x[1:]
plt.plot(x2, Annual_change_per100, label='Line Plot', color='blue', marker='o')
plt.grid(True)
plt.xlabel("Year")
plt.ylabel("Annual Change of Passenger Numbers in %")
plt.title("Demand for Zurich Airport")
plt.legend()  # Display legend
plt.figure()
```
